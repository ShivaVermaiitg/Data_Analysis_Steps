# Laptop Pricing Data Set

## Importing Required Libraries

```python
import pandas as pd
import numpy as np
```

```python
import requests

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-Coursera/laptop_pricing_dataset_base.csv"
file_name = "laptops.csv"

response = requests.get(url)

with open(file_name, 'wb') as file:
    file.write(response.content)

print(f"File downloaded and saved as {file_name}")
```

```python
df = pd.read_csv(file_name)
```

# Task-1
### Load the Dataset in pandas dataframe named 'df'

```python
df = pd.read_csv(file_name, header=None)
print(df.head())
```

# Task-2
### Add header to the dataframe

```python
# create header list
headers = ["Manufacturer", "Category", "Screen", "GPU", "OS", "CPU_core", "Screen_Size_inch", "CPU_frequency", "RAM_GB", "Storage_GB_SSD", "Weight_kg", "Price"]
df.columns = headers
print(df.head(10))
```

# Task-3

### Replace '?' with 'NaN'

```python
df.replace('?',np.nan, inplace = True)
```

### Print the data types of the dataframe columns

```python
print(df.dtypes)
```
